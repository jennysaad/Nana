```python
import numpy as np
import pandas as pd

folder = '/content/drive/MyDrive/training'  # paste your actual folder name here

# Load labels
labels = pd.read_csv(f'{folder}/train_label_mapping.csv')
print(labels.head())

# Load all EEG files
data = {}
for fname in os.listdir(folder):
    if fname.endswith('.npy'):
        subject_id = fname.replace('.npy', '')
        data[subject_id] = np.load(f'{folder}/{fname}', allow_pickle=True)

print(f'Loaded {len(data)} subjects')
print(f'Example shape: {list(data.values())[0].shape}')
```
